### Cleaning the data

Converting User Id values to Strings and Column names to Dates

```python
df = pd.DataFrame(data)
df = df.astype(float)
df = df.applymap(str)
df.columns = pd.to_datetime(df.columns)
```

### How has Daily Active Users(DAU) varied across time?


Count number of unique daily active users

```python
daily_active_users = df.nunique()-1
days = df.columns
df.nunique()
```

```python
f = plt.figure()
plt.plot(days, daily_active_users)
plt.scatter(days, daily_active_users)
plt.title("DAU Variation With Time")
plt.xlabel("Dates")
plt.ylabel("Daily Active Users")
f.set_figwidth(20)
f.set_figheight(5)
plt.show()
```

Create new dataframe with info

```python
df_info = pd.DataFrame(days, columns = ['Date'])
df_info['Month'] = days.month
df_info['Week'] = days.week
df_info['DAU'] = list(daily_active_users)
df_info
```

### What proportion of DAU is contributed by new users vs old users?

```python
new_users = []
old_users = []
for i in range(len(df.columns)):
    previous = np.arange(i)
    previous_days = df.iloc[:,previous]
    today = df.iloc[:,i]
    old = 0
    new = 0
    for user in today:
        if user == 'nan':
           continue
        if user in previous_days.values:
            old += 1
        else:
            new += 1
    old_users.append(old)         
    new_users.append(new)    
```

```python
df_info['New'] = new_users
df_info['Old'] = old_users
df_info
```

```python
total = np.add(new_users, old_users)
new_percent = np.round((np.divide(new_users, total) * 100),2)
old_percent = np.round((np.divide(old_users, total) * 100),2)
print(old_percent,"\n", new_percent)
```

```python
x = np.arange(len(days))
f = plt.figure()
plt.bar(x, old_users, width=0.4, linewidth=3)
plt.bar(x+0.4, new_users, width=0.4, linewidth=3)
plt.legend(["New users", "Old users"], loc ="upper left")
plt.xlabel("Days")
plt.ylabel("Daily Active Users")
f.set_figwidth(20)
f.set_figheight(5)
plt.show()
```

### What metrics should we track? Eg - Daily retention/ Weekly retenion/ Monthly retention, etc

Retention Rate = ((CE-CN) / CS)) X 100 

CE = This is the number of customers at the end of the period

CN = This is the number of new customers acquired during the period

CS = This is the number of customers at the start of the period

```python
daily_retention = []    
start_count = 2
days = df.columns
for day in range(0,len(days)):
    daily_retention.append(np.round(((len(df.iloc[:,day].unique())-new_users[day])/start_count * 100),2))
    start_count = len(df.iloc[:,day].unique())
print(daily_retention)
```

```python
f = plt.figure()
plt.bar(days,daily_retention)
plt.xlabel("Days")
plt.ylabel("Daily Retention (in %)")
plt.title("Daily Retention wrt Time")
f.set_figwidth(15)
f.set_figheight(5)
plt.show()
```

```python
weekly_retention = []
start_count = 2
days = df.columns
for week in range(0,len(days),7):
    weekly_retention.append(np.round(((len(df.iloc[:,week].unique())-new_users[week])/start_count * 100),2))
    if week != 90:
        start_count = len(df.iloc[:,week+1])
print(weekly_retention)
```

```python
x = np.arange(len(weekly_retention))
f = plt.figure()
plt.bar(x,weekly_retention)
plt.xlabel("Weeks")
plt.ylabel("Weekly Retention (in %)")
plt.title("Weekly Retention wrt Time")
f.set_figwidth(15)
f.set_figheight(5)
plt.show()
```

```python
monthly_retention = []
start_count = 2
days = df.columns
for month in range(0,len(days),31):
    monthly_retention.append(np.round(((len(df.iloc[:,month].unique())-new_users[month])/start_count * 100),2))
    if month != 90:
        start_count = len(df.iloc[:,month+1])
print(monthly_retention)
```

```python
months = days.month.unique()
f = plt.figure()
plt.bar(months,monthly_retention)
plt.xlabel("Months")
plt.ylabel("Monthly Retention (in %)")
plt.title("Monthly Retention wrt Time")
f.set_figwidth(15)
f.set_figheight(5)
plt.show()
```
